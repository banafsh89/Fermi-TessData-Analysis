```python
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
```

```python
filesBG=sorted(glob.glob("../NASA_ADAP_TESS_AGN/4FGLJ0601.1-7035/lygos/data/tserrflxtarg_psfn4FGLJ0601.1-7035_TESS*"))
filesSig=sorted(glob.glob("../NASA_ADAP_TESS_AGN/4FGLJ0601.1-7035/data/arrytserbdtrTESSc*"))
```

```python
for fileBG in filesBG:
    dfBG=pd.read_csv(fileBG,sep=",",header=[0])
    dfBG.columns=["t_BJD","relative_flux","E_relative_flux"]
    
    for fileSig in filesSig:
        dfSig=pd.read_csv(fileSig,sep=",",header=[0])
        print("Number of data rows in each signal file: ", dfSig.shape[0])
        dfSig.columns=["t_BJD","relative_flux","E_relative_flux"]
        
        if (dfSig["t_BJD"].min()==dfBG["t_BJD"].min()) and (dfSig["t_BJD"].max()==dfBG["t_BJD"].max()):
            break
        else:
            continue
            
    #print(fileBG,fileSig)
    
    #different selection criteria to clean the data:
    
    #1- ignore the section if more than 10% of the data is negative:
    negPer=dfSig[dfSig["relative_flux"]<0].shape[0]/dfSig.shape[0]
    if negPer > 0.1:
        print("removed file ",fileSig)
        continue
        
    #2- compare averages:
    meanSig=dfSig["relative_flux"].mean()
    stdSig=dfSig["relative_flux"].std()
    meanBG=dfBG["relative_flux"].mean()
    stdBG=dfBG["relative_flux"].std()
    
    if meanSig<=meanBG:
        print("removed due to sig({})<=BG({})".format(meanSig,meanBG))
        continue
        
    plt.errorbar(1,meanSig,yerr=stdSig, fmt=".", color="red", ecolor='lightcoral', elinewidth=2, capsize=2)
    plt.errorbar(1,meanBG,yerr=stdBG, fmt=".", color="blue", ecolor='lightblue', elinewidth=1, capsize=1)
    plt.show()
    
    
    #3- find correlation:
    dfBG["relative_flux_sub"]=dfBG["relative_flux"]-meanBG
    dfSig["relative_flux_sub"]=dfSig["relative_flux"]-meanSig
    
    corr=sum(dfBG["relative_flux_sub"]*dfSig["relative_flux_sub"])
        
    print(corr)
    fig, axs = plt.subplots(nrows=1, ncols=1, constrained_layout=False,figsize=(10, 5))
    axs.tick_params(axis='both', which='major', labelsize=14)
    axs.tick_params(axis='both', which='minor', labelsize=14)
    plt.scatter(dfSig['t_BJD'],dfSig['relative_flux'], c='red', marker=".")
    plt.scatter(dfBG['t_BJD'],dfBG['relative_flux'], c='blue', marker=".")
    plt.xlabel("time (BJD)",fontsize=14)
    plt.ylabel("Relative Flux",fontsize=14)
    plt.show()
    
```
